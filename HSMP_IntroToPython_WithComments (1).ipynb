```python
# Tutorial on Subsetting Lists in Python
# Subsetting lists refers to the process of accessing a subset of elements from a list.

# Zero Indexing
# Zero indexing is a method of numbering in which the initial element of a sequence is assigned the index 0.
# This means the first element of the list has an index of 0, the second element has an index of 1, and so on.

# Let's define a list that includes room names and their respective areas.
areas = ["hallway", 11.25, "kitchen", 18.0, "living room", 20.0, "bedroom", 10.75, "bathroom", 9.50]

# List Slicing
# In Python, list slicing allows us to access a range of elements within a list.
# The syntax is list[start:stop], where 'start' is the index of the first element to include (inclusive),
# and 'stop' is the index of the first element to exclude (exclusive).

# Print out the second element from the list
print(areas[1])  # Output: 11.25

# Print out the last element from the list
print(areas[-1])  # Output: 9.50

# In zero-indexed languages like Python, using negative indices allows you to access elements from the end of the list.
# -1 refers to the last element, -2 to the second last element, -3 to the third last element, and so on.

# Print out the area of the living room
print(areas[5])  # Output: 20.0

# Let's summarize the indices for our list:
# areas[0] -> "hallway"
# areas[1] -> 11.25
# areas[2] -> "kitchen"
# areas[3] -> 18.0
# areas[4] -> "living room"
# areas[5] -> 20.0
# areas[6] -> "bedroom"
# areas[7] -> 10.75
# areas[8] -> "bathroom"
# areas[9] -> 9.50

# Using these indices, we can access and manipulate the elements within the list as needed.
```

```python
# Tutorial on List Manipulation in Python

## Subsetting Lists
# Subsetting lists refers to accessing a subset of elements from a list using indexing or slicing.

### Zero Indexing
# Zero indexing is a method of numbering in which the initial element of a sequence is assigned the index 0.
# This means the first element of the list has an index of 0, the second element has an index of 1, and so on.

areas = ["hallway", 11.25, "kitchen", 18.0, "living room", 20.0, "bedroom", 10.75, "bathroom", 9.50]

# Print out the second element from the list
print(areas[1])  # Output: 11.25

# Print out the last element from the list
print(areas[-1])  # Output: 9.50

# In zero-indexed languages like Python, using negative indices allows you to access elements from the end of the list.
# -1 refers to the last element, -2 to the second last element, -3 to the third last element, and so on.

# Print out the area of the living room
print(areas[5])  # Output: 20.0

## List Manipulation
# You can manipulate lists by changing, adding, or removing elements.

### Changing List Elements
# You can change elements of a list by accessing the index and assigning a new value.

# Create the areas list
areas = ["hallway", 11.25, "kitchen", 18.0, "living room", 20.0, "bedroom", 10.75, "bathroom", 9.50]

# Correct the bathroom area
areas[-1] = 10.50

# Change "living room" to "chill zone"
areas[4] = "chill zone"

### Adding and Removing Elements
# You can add and remove elements from a list using various methods.

# Create the updated areas list
areas = ["hallway", 11.25, "kitchen", 18.0, "chill zone", 20.0, "bedroom", 10.75, "bathroom", 10.50]

# Add poolhouse data to areas, creating a new list areas_1
areas_1 = areas + ["poolhouse", 24.5]

# Delete the element at index 2 ("kitchen")
del areas[2]
print(areas)  # Output: ['hallway', 11.25, 18.0, 'chill zone', 20.0, 'bedroom', 10.75, 'bathroom', 10.50]

# Add garage data to areas_1, creating a new list areas_2
areas_2 = areas_1 + ["garage", 15.45]

### Common List Methods
# Python lists come with several built-in methods for common operations.

# Initial list for demonstration
my_list = [10, 20, 30, 40, 50]

# 1. append() - Add an element to the end of the list
my_list.append(60)
print("After append(60):", my_list)  # Output: [10, 20, 30, 40, 50, 60]

# 2. extend() - Extend the list by appending elements from an iterable
my_list.extend([70, 80])
print("After extend([70, 80]):", my_list)  # Output: [10, 20, 30, 40, 50, 60, 70, 80]

# 3. insert() - Insert an element at a specified position
my_list.insert(2, 25)  # Insert 25 at index 2
print("After insert(2, 25):", my_list)  # Output: [10, 20, 25, 30, 40, 50, 60, 70, 80]

# 4. remove() - Remove the first occurrence of an element
my_list.remove(30)
print("After remove(30):", my_list)  # Output: [10, 20, 25, 40, 50, 60, 70, 80]

# 5. pop() - Remove and return the element at a specified position (or the last element if not specified)
popped_element = my_list.pop()
print("After pop():", my_list)  # Output: [10, 20, 25, 40, 50, 60, 70]
print("Popped element:", popped_element)  # Output: 80

# 6. index() - Return the index of the first occurrence of an element
index_of_50 = my_list.index(50)
print("Index of 50:", index_of_50)  # Output: 4

# 7. count() - Return the number of occurrences of an element
count_of_20 = my_list.count(20)
print("Count of 20:", count_of_20)  # Output: 1

# 8. sort() - Sort the list in ascending order
my_list.sort()
print("After sort():", my_list)  # Output: [10, 20, 25, 40, 50, 60, 70]

# 9. reverse() - Reverse the elements of the list
my_list.reverse()
print("After reverse():", my_list)  # Output: [70, 60, 50, 40, 25, 20, 10]

# 10. copy() - Return a shallow copy of the list
copied_list = my_list.copy()
print("Copied list:", copied_list)  # Output: [70, 60, 50, 40, 25, 20, 10]

# 11. clear() - Remove all elements from the list
my_list.clear()
print("After clear():", my_list)  # Output: []
```

```python
#if Statement
# The if statement is used to test a condition. If the condition is true, the code block following the if statement is executed.

x = 10
if x > 5:
    print("x is greater than 5")

#elif Statement
#The elif (short for "else if") statement is used to check another condition if the previous if condition is false. 
#You can have multiple elif statements to check multiple conditions.
x = 10
if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but less than or equal to 15")

#else Statement
#The else statement is used to execute a code block if none of the previous conditions (if and elif) are true.
x = 10
if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but less than or equal to 15")
else:
    print("x is 5 or less")
```

```python
# I dont think we had time for this in our lesson, but test this out! 

# Create an interactive BMI Calculator 

# Function to calculate BMI
def calculate_bmi(weight, height):
    # Calculate BMI using the formula: weight / (height^2)
    return weight / (height ** 2)

# Function to determine the BMI category
def get_bmi_category(bmi):
    # Determine the category based on the BMI value
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"

# Main function to run the BMI calculator
def main():
    # Ask the user for their weight in kilograms
    weight = float(input("Enter your weight in kilograms: "))
    # Ask the user for their height in meters
    height = float(input("Enter your height in meters: "))
    
    # Calculate the BMI
    bmi = calculate_bmi(weight, height)
    
    # Determine the BMI category
    category = get_bmi_category(bmi)
    
    # Print the results
    print(f"Your BMI is: {bmi:.2f}")
    print(f"You are classified as: {category}")

main()
```
